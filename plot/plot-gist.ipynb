```python
import pandas as pd

annoy = pd.read_csv('../results/annoy-gist.csv', index_col=0)
annoy['method'] = 'annoy'
annoy.head()
```

```python
annoy.plot(x="recall", y="QPS")
```

```python
LSH = pd.read_csv('../results/lsh-gist.csv', index_col=0)
LSH['method'] = "LSH"

PQ = pd.read_csv('../results/pq-gist.csv', index_col=0)
PQ['method'] = "PQ"

HNSW = pd.read_csv('../results/hnsw-gist.csv', index_col=0)
HNSW['method'] = "HNSW"

df = pd.concat([annoy, LSH, PQ, HNSW])
df.head(n=10)
```

```python
import seaborn as sns

sns.lineplot(data=df, x='recall', y='QPS', hue='method')
```

```python
sns.lineplot(data=df, x='recall', y='build_time', hue='method')
```
